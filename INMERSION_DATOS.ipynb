```python
# Empezamos separando el simbolo $ en una columna y los valores en millones de pesos colombianos en otra

valor = inmuebles.Valor.str.split(expand = True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

```python
# Ahora quitamos los puntos y los reemplazamos

inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex = True)
```

```python
# Para hacerlo más legible, crearemos una nueva columna 'Precio_Millon' donde los valores irán abreviados para poder trabajar en ellos de mejor manera

inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

```python
# Usamos la función describe para revisar que tal nos quedaron los datos y de una vez ir sacando conclusiones

inmuebles.describe()
```

```python
# Cambiamos los números para que solo tengan dos decimales, así es más legible

pd.set_option('display.precision', 2)
pd.set_option('display.float_format',lambda x: '%.2f' % x)
inmuebles.describe()
```

```python
# Ahora empezaremos nuestro modelo de ML, importamos las librerías para realizar este modelo

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Se busca predecir el Precio en millones, utilizando en este caso el codigo UPZ para ver como ingluye en el precio la locacion de los inmuebles

X = datos_ml[['COD_UPZ_GRUPO']]
y = datos_ml["Precio_Millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
```
